데이터프레임 데이터 변환
1. 특정 컬럼만 선택해서 데이터프레임 만들기.
2. 특정 컬럼에 없는 데이터 삭제. (Confirmed 컬럼을 가지고 분석해야하니까 Confirmed 값이 null인건 삭제)
3. 특정 컬럼의 데이터 타입 변경.

```python
doc = pd.read_csv(PATH + "01-22-2020.csv", encoding='utf-8-sig')
try:
    doc = doc[['Province_State', 'Country_Region', 'Confirmed']]  #1.특정 컬럼만 선택해서 데이터 프레임 만듬.
except:
    doc = doc[['Province/State', 'Country/Region', 'Confirmed']]  #1.특정 컬럼만 선택해서 데이터 프레임 만듬.
    doc.columns = ['Province_State', 'Country_Region', 'Confirmed']
doc = doc.dropna(subset=['Confirmed'])  #2. 특정컬럼에 없는 데이터 삭제하기
doc = doc.astype({'Confirmed' : 'int64'}) #3. 특정 컬럼 데이터 타입 변경
doc.head()
```


컬럼값 변경하기

-Country_Region 국가명이 다양한 경우가 많음.

-json 파일로 국가명을 일관되게 변경.


```python
import json

with open('COVID-19-master/csse_covid_19_data/country_convert.json', 'r', encoding = 'utf-8-sig') as json_file:
    json_data = json.load(json_file)
    print(json_data.keys())
```


apply() 함수를 사용해서, 국가 컬럼값 변경.
-Country_Region 이라는 컬럼값을 확인해서, 국가명이 다르게 기재되어 있는 경우에만, 지정한 국가명으로 변경


```python
def func(row):
    if row['Country_Region'] in json_data:
        print(json_data.items())
        row['Country_Region'] = json_data[row['Country_Region']]
    return row
```

```python
doc = doc.apply(func, axis=1)
doc.head()
```
